> ## 1) 'Camel.npy'

* numpy Data

```python
import numpy as np

CR = np.load('Camel.npy')

CR.shape
```

```python
CR = CR.reshape(len(CR), 28, 28, 1).astype('float32')

CR.shape
```

```python
np.set_printoptions(linewidth = 180)

CC = np.squeeze(CR)

CC[0]
```

```python
import matplotlib.pyplot as plt

plt.imshow(CC[0], cmap = 'gray')
plt.show()
```
